### Box Plot เป็นGraph ซึ่งสามารถทำให้เห็นภาพของข้อมูลได้อย่างชัดเจน

```python
from matplotlib import pyplot as plt
```

```python
import pandas
```

```python
irisdata = pandas.read_csv('C:\\Users\\KD\\Documents\\gatk\\python_dataviz\\iris.data',header=None)
irisdata.head()
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

### import ไลบรารี matplotlib เพื่อเรียกใช้งานฟังก์ชันสำหรับพล็อตกราฟกล่อง

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [12, 12]
plt.subplot(1,4,1)  # 1 เเถว 4 คอลัมน์ ตำเเหน่งที่ 1
plt.boxplot(setosa_np[:,0])
plt.subplot(1,4,2)
plt.boxplot(setosa_np[:,1])
plt.subplot(1,4,3)
plt.boxplot(setosa_np[:,2])
plt.subplot(1,4,4)
plt.boxplot(setosa_np[:,3])
plt.show()
```

```python
plt.boxplot(setosa_np)  ### plot รวมกันทั้งหมด
plt.title('Setosa');
```

#### import folium เพื่อใช้สำหรับการทำแผนที่

```python
import folium
import pandas as pd
```

```python
data_to_plot = Debt_df_debt
```

```python
data_to_plot
```

```python
data_to_plot.head()
```

#### นำไฟล์เข้า

```python
country_geo = ('C:\\Users\\KD\\Documents\\thai.json')
```

```python
new_col=list()
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province']=new_col
```

#### ทำการกำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

```python
map = folium.Map(location=[10,100], zoom_start=5)
```

```python
map
```

#### ทำการลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือนถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมีค่ามากจะเป็นสีม่วงเข้ม ส่วนจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลืองตามแถบรูปข้างบน

```python
map.choropleth(geo_data=country_geo, data=data_to_plot,
              columns=['Region and province',2549],
              key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7, line_opacity=0.2,
              legend_name='Average Debt')
```

```python
map.save('plot_data.html')
```
